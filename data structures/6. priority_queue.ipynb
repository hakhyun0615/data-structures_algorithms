우선순위 큐 (←=←): 가장 높은 우선순위의 항목을 삭제 및 반환하는 큐
- {스택, 큐} ⊂ 우선순위 큐
- 가장 높은 우선순위: ex) 가장 작은 항목
- 삽입: enqueue()
- 삭제 및 반환: dequeue()

물리적 구현
- 파이썬 리스트
    - 비정렬
        - 삽입(append): **O(1)** 
        - 탐색(O(N)) 후 삭제 및 반환(자리이동,O(N)): **O(N)**
    - 내림차순 정렬
        - 탐색(O(N)) 후 삽입(자리이동,O(N)): **O(N)**
        - 삭제 및 반환(pop): **O(1)**
- 정렬된 단순 연결 리스트 (사용 x)
    - 오름차순 정렬
        - 탐색(O(N)) 후 삽입: **O(N)**
        - 삭제 및 반환: **O(1)**
- **힙**
    - 삽입: **O(logN)**
    - 삭제 및 반환: **O(logN)**
